**Brief Introduction to Titanic Dataset:**

The Titanic dataset typically includes the following columns:

1. **PassengerId:** Unique identifier for each passenger.
2. **Survived:** Survival status of the passenger (0 = No, 1 = Yes).
3. **Pclass:** Ticket class indicating the socio-economic status of the passenger (1 = 1st class, 2 = 2nd class, 3 = 3rd class).
4. **Name:** Name of the passenger.
5. **Sex:** Gender of the passenger.
6. **Age:** Age of the passenger in years.
7. **SibSp:** Number of siblings/spouses aboard the Titanic.
8. **Parch:** Number of parents/children aboard the Titanic.
9. **Ticket:** Ticket number.
10. **Fare:** Passenger fare.
11. **Cabin:** Cabin number.
12. **Embarked:** Port of embarkation (C = Cherbourg, Q = Queenstown, S = Southampton).

Loop Explanation

Imagine you have a task to count the number of apples in a basket. You start by picking up the first apple, checking if it's there, then moving on to the next one, and so on, until you've checked all the apples in the basket. This process of picking up each apple, checking it, and moving to the next one is similar to what a loop does in programming.

In programming, a loop is like a repeated set of instructions that the computer follows until a certain condition is met. It's like telling the computer, "Hey, do this thing again and again until you're told to stop."

There are mainly two types of loops: for loops and while loops.

For loop:
Imagine you have a list of tasks to do. You go through the list, doing each task one by one until you've completed all of them. That's what a for loop does. It helps you repeat a certain block of code a fixed number of times.For example, if you have 10 apples in the basket, you'll repeat the process of picking up each apple and checking it 10 times.
While loop:
Now, imagine you have a task to do, but you're not sure how many times you'll have to do it. You keep doing the task until a specific condition is met. That's what a while loop does. It helps you repeat a certain block of code as long as a condition is true.For example, you keep picking up apples from the basket and checking them until you've checked all the apples (i.e., until the basket is empty).
In programming, loops are powerful tools that help us automate repetitive tasks and make our code more efficient.

Count of Missing Values: Calculate and display the count of missing values in each column.


```python
# Count of missing values
missing_values_count = df.isnull().sum()
print("Count of Missing Values:")
print(missing_values_count)
```

```python
# Verify if there are any missing values left
print("Count of Missing Values:")
print(df.isnull().sum())
```

Count of Survivors: Calculate the count of passengers who survived and who didn't survive.

```python
# Count of survivors
survived_count = df['Survived'].value_counts()

# Extract counts of survivors and non-survivors
survived = survived_count[1]
not_survived = survived_count[0]

# Print counts
print("Count of Survivors:", survived)
print("Count of Non-Survivors:", not_survived)
```

```python
# Data for pie chart
labels = ['Survived', 'Not Survived']
sizes = [survived, not_survived]
colors = ['lightgreen', 'lightcoral']


# Create pie chart
plt.figure(figsize=(8, 6))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', shadow=False, startangle=140)
plt.title('Survival Status')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

Survival Rate by Class: Calculate and visualize the survival rate by passenger class.

```python
# Survival Rate by Class
survival_rate_by_class = df.groupby('Pclass')['Survived'].mean()

# Convert survival rates to percentages rounded to two decimal places
survival_rate_by_class_percent = (survival_rate_by_class * 100).round(2)

# Print survival rates as percentages with percent sign
print("Survival Rate by Class:")
for pclass, survival_rate in survival_rate_by_class_percent.items():
    print(f"Pclass {pclass}: {survival_rate}%")


# Visualize survival rate by class
survival_rate_by_class.plot(kind='bar', color='green')
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate')
plt.title('Survival Rate by Class')
plt.xticks(rotation=0)
plt.show()
```

Survival Rate by Sex: Calculate and visualize the survival rate by gender.


```python
# Count survivors and non-survivors by sex
survival_count_by_sex = df.groupby(['Sex', 'Survived']).size()

# Print survival count by sex
print("Survival Count by Sex:")
print(survival_count_by_sex)

# Visualize survival rate by sex
survival_rate_by_sex = df.groupby('Sex')['Survived'].size()
survival_rate_by_sex.plot(kind='bar', color='purple')
plt.xlabel('Sex')
plt.ylabel('Survival Rate')
plt.title('Survival Rate by Sex')
plt.xticks(rotation=0)
plt.show()
```

Survival Rate by Age Group: Calculate and visualize the survival rate by age group.


```python
# Bin ages into groups
age_bins = [0, 18, 30, 50, 100]
age_labels = ['0-18', '19-30', '31-50', '51+']
df['Age_Group'] = pd.cut(df['Age'], bins=age_bins, labels=age_labels)

# Survival rate by age group
survival_rate_by_age_group = df.groupby('Age_Group')['Survived'].size()
print("Survival Rate by Age Group:")
print(survival_rate_by_age_group)

# Visualize survival rate by age group
survival_rate_by_age_group.plot(kind='bar', color='blue')
plt.xlabel('Age Group')
plt.ylabel('Survival Rate')
plt.title('Survival Rate by Age Group')
plt.xticks(rotation=0)
plt.show()
```

Survival Rate by Embarkation Port: Calculate and visualize the survival rate by embarkation port.


```python
# Survival rate by embarkation port
survival_rate_by_embarkation = df.groupby('Embarked')['Survived'].size()
print("Survival Rate by Embarkation Port:")
print(survival_rate_by_embarkation)

# Visualize survival rate by embarkation port
survival_rate_by_embarkation.plot(kind='bar', color='red')
plt.xlabel('Embarkation Port')
plt.ylabel('Survival Rate')
plt.title('Survival Rate by Embarkation Port')
plt.xticks(rotation=0)
plt.show()
```

Survival Rate by Family Size: Calculate and visualize the survival rate by family size (combining SibSp and Parch).


```python
# Calculate family size
df['Family_Size'] = df['SibSp'] + df['Parch'] + 1

# Survival rate by family size
survival_rate_by_family_size = df.groupby('Family_Size')['Survived'].size()
print("Survival Rate by Family Size:")
print(survival_rate_by_family_size)

# Visualize survival rate by family size
survival_rate_by_family_size.plot(kind='bar', color='brown')
plt.xlabel('Family Size')
plt.ylabel('Survival Rate')
plt.title('Survival Rate by Family Size')
plt.xticks(rotation=0)
plt.show()
```
